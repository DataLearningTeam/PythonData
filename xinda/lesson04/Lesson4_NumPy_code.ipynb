## 2. 索引和切片

选取数组中的元素或者是数据的子集

使用[ ]运算符对数进行切片和索引等操作

### 2-1. 一维数组
* 一维数组的索引和切片，与Python列表差不多

```python
# 创建一个一维的数组
arr = np.arange(10)
arr
```

```python
# 一维数组的索引
arr[3]
```

```python
# 一维数组的切片
arr[3:7]
```

* 数组切片是原始数组的视图，也就是说数据不会被复制，任何修改都会影响到原有的数组上。

```python
arr2 = arr[3:7]
arr2[0] = 13
arr              # arr的元素被改变了
```

* 如果要对数组的切片进行复制，需要使用`.copy()`方法

```python
arr = np.arange(10)
arr3 = arr[3:7].copy()   # 使用.copy()
arr3[0] = 13
arr                      # arr的元素没有被改变
```

### 2-2. 二维数组

```python
# 用一维数组来生成二维数组
mat = np.arange(1,17)
mat.shape = (4, 4)
mat
```

* 获取二维数组的某一行 

```python
# 对于二维数组，每个索引对应的是一个一维数组！
mat[0]  # 对应第0行
```

* 获取二维数组的行

    * mat[起始行:终结行] ：从起始行开始，但不包括终结行
    * mat[起始行:] ： 从起始行到最后一行
    * mat[: 终结行] ： 从第0行到终结行但不包括终结行

```python
# 类似于列表切片，查看第二和第三行
# ndarray的切片是沿着行方向的
# 和列表切片一样，第三行是不包括的
mat[1:3] #等价于mat[[1,2]]
```

* 获取二维数组的行和列

下面的矩阵给出了数组里的每一个元素的索引


\begin{bmatrix}
    (0,0) & (0,1) & (0,2) & (0,3) \\
    (1,0) & (1,1) & (1,2) & (1,3) \\
    (2,0) & (2,1) & (2,2) & (2,3) \\
    (3,0) & (3,1) & (3,2) & (3,3)
\end{bmatrix}

* 查看某一个元素，有两种方式

```python
# 先获取第一行，再查找第三列
mat[1][3] 
```

```python
# 用一个逗号来隔开行和列的索引
mat[1,3]
```

* 索引和切片的混用

方括号第一个参数指定要取的行位置，第二个参数指定要取得列位置，中间使用逗号隔开

```python
# 取出第0行到第2行，第1列
mat[0:3, 1]
```

```python
# 可以同时在行和列同时做切片
# 取出第1行到最后一行，第2列到第3列
mat[2:, 2:]
```

## 4. 布尔型索引
* 之前的切片索引方法是通过指定行或列的位置来取一个数组的一部分
* 顾名思义，布尔索引是通过一个布尔型数组来确定所需要的行或者列的位置
* 当行或列位置对应的布尔索引为True的时候，我们会保留这个行或列。反之，当对应的布尔索引为False的时候，则不选取。

*例1：获取评分大于5的红酒数据*

```python
# 第2列数据代表质量评分
wine[:, 2] > 5   
```

```python
wine[wine[:, 2] > 5]
```

*例2：获取评分大于5，且酒精度大于等于10的红酒数据*

```python
# 第1列代表酒精度，第2列代表评分
# 使用布尔运算符 (& , | ) 来组合多个布尔条件
mask = (wine[:,2] > 5) & (wine[:,1] >= 10)
mask
```

```python
wine[mask]
```

## 5. 数组的运算

### 5-1. 数组与单个数之间的运算

```python
# 每个元素都除以2
mat / 2
```

```python
# 每个元素都平方
mat ** 2 
```

```python
(mat + 100) / 100
```

*例：将红酒评分数据转换成百分制，即原评分乘以10*

```python
# 获取评分数据的数组
score = wine[:, 2]
score
```

```python
score * 10
```

* Numpy 数组的任何算数运算，都会将运算运用到**元素**级别。
* 这样的矢量化运算是NumPy 数组的优势， Python列表要实现这一操作就需要编写循环。

```python
# 将上述评分数据转化成Python列表
score_list = list(score)
score_list
```

```python
# 直接将列表乘以10，并不能得到类似数组的计算结果，而是把列表复制了10遍
print(score_list * 10)
```

```python
# 方法一：使用循环将列表的每一个元素乘以10
new_list = list()
for s in score_list:
    new_list.append(s * 10)
print(new_list)
```

```python
# 方法二：使用列表解析
[ s * 10 for s in score_list]
```

### 5-2. 数组与数组之间的运算
* 两个行列数相符的数列可以进行运算
* 运算操作将应用到数列中对应的每一个元素

```python
# 两个数组相加
mat + mat
```

```python
np.add(mat, mat)
```

```python
# 两个数组相乘， 也可以使用np.multiply函数实现
# 对应元素之间的乘法，不同于线性代数中的矩阵相乘
mat * mat  
```

*例：将每一行的红酒数据与第0行数据进行比较，即计算与第0行数据的差值*

```python
# 注意，这里两个数组的大小不同
# 大小不同的数组之间的运算叫做广播
wine - wine[0]
```

* NumPy 数组比 Python 列表效率更高

数组和列表都有max函数可以找到最大值，我们来看下他们的运算速度。

```python
# 先生成10000个随机数，然后需要找到它们中的最大值
random_arr = np.random.rand(10000)
random_list = list(random_arr)
```

```python
# 直接使用Python的max函数
%timeit max(random_list)
```

```python
# 使用NumPY中的max函数
%timeit np.max(random_arr)
```

*例5：生成一组新的数据，如果评分大于5，其值为‘Good’；如果评分小于等于5，其值为‘Bad'*

```python
# 使用循环和判断语句
new = list()
for s in wine[:,2] :
    if s > 5 :
        new.append('Good')
    else:
        new.append('Bad')

new 
```

```python
# 使用列表解析
[('Good' if s > 5 else 'Bad') for s in wine[:,2]]
```

```python
# 使用np.where()函数
np.where(wine[:,2] > 5, 'Good', 'Bad')
```
